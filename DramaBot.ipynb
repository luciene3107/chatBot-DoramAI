```python
import requests

API_KEY = "f1b3f1fb4eb44c118840555720ef8ccc"

def buscar_dorama(nome_dorama):

    url = "https://api.themoviedb.org/3/search/tv"

    params = {
        "api_key": API_KEY,
        "query": nome_dorama,
        "language": "pt-BR"
    }

    response = requests.get(url, params=params)

    data = response.json()

    if data["results"]:
        dorama = data["results"][0]

        return {
            "titulo": dorama.get("name"),
            "descricao": dorama.get("overview"),
            "nota": dorama.get("vote_average"),
            "data": dorama.get("first_air_date"),
            "generos": dorama.get("genre_ids")
        }

    return None
```

```python
def vibe_dorama(nota):

    if nota >= 8:
        return "muito amado pelos fãs 😭✨"

    elif nota >= 7:
        return "bem popular entre fãs de dorama"

    else:
        return "mais nichado, mas pode surpreender"
```

```python
import os
import gradio as gr
from groq import Groq

# Resgata a chave de forma segura sem expor no texto do notebook
client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

MEMORIA_SESSAO = {
    "ultimo_dorama": None
}

def responder(mensagem, historico):
    print("\n=== NOVA INTERAÇÃO ===")

    saudacoes = ["oi", "olá", "ola", "bom dia", "boa tarde", "boa noite", "tudo bem", "oie"]
    mensagem_limpa = mensagem.lower().strip()

    info_dorama = None
    if mensagem_limpa not in saudacoes:
        info_dorama = buscar_dorama(mensagem)

    # Lógica de atualização da memória
    if info_dorama:
        MEMORIA_SESSAO["ultimo_dorama"] = info_dorama
        print(f">> NOVO DORAMA ENCONTRADO NO BANCO: {MEMORIA_SESSAO['ultimo_dorama']['titulo']}")
    else:
        # Ampliamos os indicadores para cobrir "me fale sobre", "conhece o", "indique", etc.
        indicadores_novo_dorama = ["sobre", "dorama", "assistir", "conhece", "fale de", "fale sobre", "recomend", "indica", "nome"]
        usuario_tentou_mudar = any(termo in mensagem.lower() for termo in indicadores_novo_dorama) or len(mensagem.split()) <= 3

        if usuario_tentou_mudar and mensagem_limpa not in saudacoes:
            MEMORIA_SESSAO["ultimo_dorama"] = None
            print(">> USUÁRIO CITOU UM ASSUNTO/DORAMA FORA DO BANCO LOCAL. Abrindo espaço para o Groq responder.")
        else:
            print(f">> MANTENDO CONTEXTO ANTERIOR: {MEMORIA_SESSAO['ultimo_dorama']['titulo'] if MEMORIA_SESSAO['ultimo_dorama'] else 'Nenhum'}")

    # --- NOVO FILTRO DE BLOQUEIO INTELIGENTE ---
    temas_dorama = [
        "dorama", "kdrama", "coreano", "netflix", "romance", "série", "serie",
        "ator", "atriz", "episódio", "episodio", "dorameiro", "dorameira",
        "sinopse", "elenco", "protagonista", "personagem", "assistir", "beijo",
        "fale", "sobre", "indica", "recomenda", "quem é", "quem e", "vibe"
    ]

    usuario_falou_de_dorama = any(palavra in mensagem.lower() for palavra in temas_dorama)
    eh_saudacao = mensagem_limpa in saudacoes

    # SÓ bloqueia de cara se for algo absurdamente fora de contexto (ex: "o que é uma variavel?")
    # Se a pessoa disser "me fale sobre...", o termo "sobre" ou "fale" vai deixar passar para a IA responder!
    if not usuario_falou_de_dorama and not eh_saudacao and not MEMORIA_SESSAO["ultimo_dorama"]:
        print(">> BLOQUEIO PREVENTIVO ACIONADO.")
        texto_bloqueio = "Eu fui criada apenas para conversar sobre doramas 😄💕"
        return "", historico + [{"role": "user", "content": mensagem}, {"role": "assistant", "content": texto_bloqueio}]

    # Montagem do prompt para o Groq
    mensagens = [{"role": "system", "content": SYSTEM_PROMPT}]

    if MEMORIA_SESSAO["ultimo_dorama"]:
        dorama_atual = MEMORIA_SESSAO["ultimo_dorama"]
        vibe = vibe_dorama(dorama_atual['nota'])
        contexto_sistema = f"""
[DORAMA EM FOCO NO BANCO LOCAL]
Título: {dorama_atual['titulo']}
Descrição: {dorama_atual['descricao']}
Nota: {dorama_atual['nota']}
Data: {dorama_atual['data']}
Vibe: {vibe}
Instrução: Use os dados acima se o usuário perguntar especificamente sobre as informações deste dorama.
--------------------------------------------------"""
        mensagens.append({"role": "system", "content": contexto_sistema})
    else:
        # Se o dorama não está no banco (como visto no seu último print), damos passe livre amigável para o Groq usar o conhecimento dele
        contexto_geral = "\nDiretriz: O dorama citado pelo usuário não está no catálogo local da aplicação. Use seu conhecimento geral do Llama para trazer a sinopse ou informações reais sobre ele em português de forma super fofa!"
        mensagens[0]["content"] += contexto_geral

    # Histórico de conversas
    if historico:
        for turno in historico:
            mensagens.append({"role": turno["role"], "content": turno["content"]})

    mensagens.append({"role": "user", "content": mensagem})

    # Chamada ao Groq
    resposta = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=mensagens,
        max_tokens=1024
    )

    texto_resposta = resposta.choices[0].message.content

    # --- SOFT-BLOCK PÓS-RESPOSTA ---
    # Se a pergunta passou pelo primeiro filtro (ex: "o que é...") mas a IA respondeu sobre código ou finanças:
    palavras_bloqueadas = ["futebol", "copa do mundo", "messi", "argentina", "política", "variável", "variable", "python", "código", "programação"]
    if any(p in texto_resposta.lower() for p in palavras_bloqueadas) or any(p in mensagem.lower() for p in ["variável", "programação"]):
         texto_resposta = "Eu adoro conversar e tirar dúvidas, mas meu mundinho são apenas os doramas! Que tal voltarmos a falar sobre romances e séries coreanas? 😉💕"

    # Atualiza a tela
    historico_atualizado = historico + [
        {"role": "user", "content": mensagem},
        {"role": "assistant", "content": texto_resposta}
    ]

    return "", historico_atualizado


# --- INTERFACE COM GRADIO BLOCKS ---
with gr.Blocks(title="🎬 DoramAI") as demo:
    gr.Markdown("# 🎬 DoramAI\n### Sua especialista em doramas 💕")

    chatbot = gr.Chatbot(type="messages", label="Conversa")

    with gr.Row():
        txt_input = gr.Textbox(
            show_label=False,
            placeholder="Digite aqui sobre o dorama...",
            scale=9
        )
        btn_enviar = gr.Button("▶️", scale=1)

    # Configuração simples: enviamos o input e o chatbot, e recolhemos o input limpo e o chatbot atualizado
    txt_input.submit(responder, inputs=[txt_input, chatbot], outputs=[txt_input, chatbot])
    btn_enviar.click(responder, inputs=[txt_input, chatbot], outputs=[txt_input, chatbot])

if __name__ == "__main__":
    demo.launch(debug=True)
```
